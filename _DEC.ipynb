```python
import keras_dec
from keras_dec import DeepEmbeddingClustering
from keras.datasets import mnist

import numpy as np
```

```python
def preproc(X):
    # 1/d * ||x_i||2**2 = 1.0
    return (X.T / X.mean(1)).T

(X_train, y_train), (X_test, y_test) = mnist.load_data()
X_train = np.asarray([x.flatten() for x in X_train], dtype='float32')
X_test = np.asarray([x.flatten() for x in X_test], dtype='float32')

X_train = preproc(X_train)
X_test = preproc(X_test)
```

c = DeepEmbeddingClustering(n_clusters=10, input_dim=784)
c.initialize(X_train, nb_epoch=20, verbose=2, save_autoencoder=True)

```python
c = DeepEmbeddingClustering(n_clusters=10, input_dim=784, pretrained_weights='autoencoder.h5')
c.initialize(X_train, nb_epoch=20, verbose=2)
```

```python
c.cluster(X_train, y=y_train, tol=.01, update_interval=None, iter_max=1e6)
```

```python
from sklearn.decomposition import PCA
z = c.encoder.predict(X_train)
pca = PCA(n_components=2).fit(z)
z_2d = pca.transform(z)
clust_2d = pca.transform(c.cluster_centres)
```
